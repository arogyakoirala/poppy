```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
DATA_DIR = "../../data"
```

```python
aoi = gpd.read_file('/data/tmp/arogya/data/inputs/districts/districts/1007.gpkg')
parent_best_dates = gpd.read_file(f"{DATA_DIR}/interim/parent_best_dates.gpkg")
parent=gpd.read_file(f"{DATA_DIR}/interim/parent.gpkg")
centroids=gpd.read_file(f"{DATA_DIR}/interim/modis/centroids.gpkg")
child=gpd.read_file(f"{DATA_DIR}/interim/child.gpkg")
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,10), dpi=100)
aoi.boundary.plot(ax=ax)
parent_best_dates.boundary.plot(color="red", ax=ax)
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,10), dpi=100)
ax=ax.flatten()
aoi.boundary.plot(ax=ax[0], color='red')
parent_best_dates.boundary.plot(color="red", ax=ax[0])
centroids.plot(color="lightblue", ax=ax[0], markersize=0.1)


aoi.boundary.plot(ax=ax[1])
parent_best_dates.boundary.plot(color="red", ax=ax[1])
centroids.plot(color="black", ax=ax[1], markersize=0.1)
# child.boundary.plot(ax=ax[1], linewidth=0.2, color='black')
ax[1].set_ylim([35.1, 35.2])
ax[1].set_xlim([71.1, 71.27])
```

```python
fig,ax = plt.subplots(1,1,figsize=(10,10), dpi=100)
aoi.boundary.plot(ax=ax)
parent_best_dates.boundary.plot(color="red", ax=ax)
centroids.plot(color="black", ax=ax, markersize=0.1)
child.boundary.plot(ax=ax, linewidth=0.2, color='black')
ax.set_ylim([35.1, 35.2])
ax.set_xlim([71.1, 71.27])
```
