```python
import os
import geopandas as gpd
import pandas as pd

districts_dir = "/data/tmp/arogya/data/inputs/districts/districts/"
districts = [file.split(".gpkg")[0] for file in os.listdir(districts_dir)]
len(districts)
areas = []
for dist in districts:
    dist = gpd.read_file(f"{districts_dir}/{dist}.gpkg").to_crs("epsg:3857")
    areas.append({"DISTID": dist["DISTID"][0], 'DIST_34_NA': dist['DIST_34_NA'][0], "area":dist.area[0] })
areas = pd.DataFrame(areas)
```

```python
areas = areas.sort_values('area', ascending=True)
list(areas['DISTID'])
```

```python
areas['perc'] = areas['area']/sum(areas['area'])
```
